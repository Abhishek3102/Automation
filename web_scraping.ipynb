<p>Injection attacks happen when untrusted data is sent to a code interpreter through a form input or some other data submission to a web application. For example, an attacker could enter SQL database code into a form that expects a plaintext username. If that form input is not properly secured, this would result in that SQL code being executed. This is known as an <a href="/learning/security/threats/sql-injection/">SQL injection attack</a>.</p>
<p>Injection attacks can be prevented by validating and/or sanitizing user-submitted data. (Validation means rejecting suspicious-looking data, while sanitization refers to cleaning up the suspicious-looking parts of the data.) In addition, a database admin can set controls to minimize the amount of information an injection attack can expose.</p>

WEB SCRAPING USING SELENIUM AND EDGE DRIVER



```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service

website = "https://www.manchestereveningnews.co.uk/all-about/manchester-united-fc"
path = "D:\Edge Driver\msedgedriver.exe"
service = Service(executable_path=path)
options = webdriver.EdgeOptions()
options.add_argument('headless')
driver = webdriver.Edge(service=service,options=options)
# driver = webdriver.Edge(service=service)
driver.get(website)

# driver.quit()
```

```python
containers = driver.find_elements(by="xpath", value='//div[@class="teaser-text"]')
```

```python
for container in containers:
    title = container.find_element(by="xpath", value='//div[@class="teaser-text"]').text
    print(title)
```

```python
containers = driver.find_elements(by="xpath", value='//div[@class="teaser-text"]/a[@class="headline"]')
for container in containers:
    title = container.text
    print(title)

```

```python
containers = driver.find_elements(by="xpath", value='//div[@class="teaser-text"]/a[@class="headline"]')
for i, container in enumerate(containers, start=1):
    title = container.text
    print(f"{i}. {title}\n")
```
